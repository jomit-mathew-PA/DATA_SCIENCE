```python
#step1 import all the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing  import StandardScaler,PolynomialFeatures
```

```python
# get the data
df=pd.read_csv("Walmart_Sales.csv")
print(df)
df.head()
```

```python
df.info()
```

```python
# drop the unwanted columns
df.drop(columns=["Date"],inplace=True)
print(df)
```

```python
model=LinearRegression()
```

```python
x=df[["Fuel_Price","Temperature","CPI","Holiday_Flag","Unemployment",'Store']]
y=df["Weekly_Sales"]
# split the data
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
# create an object for standardscaler
std=StandardScaler()
# using the fit_transform
std_train=std.fit_transform(xtrain)
# by using the polynomial features
poly=PolynomialFeatures(degree=5)
xpoly=poly.fit_transform(std_train)
model=LinearRegression()
model.fit(xpoly,ytrain)
# testing the data
xtest_scale=std.transform(xtest)
xtestpoly=poly.transform(xtest_scale)
# predicting the value
ypred=model.predict(xtestpoly)
r2_score(ytest,ypred)


```
